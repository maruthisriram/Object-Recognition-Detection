```python
import numpy as np
import pandas as pd 

import os
import random

import torchvision.models as models
import torchvision.transforms as transforms # transform data
from PIL import Image
```

```python
classes = os.listdir('../input/yikes-spiders-15-species/train')
print(classes)
```

```python
train = []
test = []
```

```python
preprocess = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
```

```python
for i in range(len(classes)):
    image_path = os.listdir('../input/yikes-spiders-15-species/train/' + classes[i] + '/')
    for j in range(104):
        image = Image.open('../input/yikes-spiders-15-species/train/' + classes[i] + '/' + image_path[j])
        input_image = preprocess(image).unsqueeze(0)
        train.append([input_image, i])
```

```python
for i in range(len(classes)):
    image_path = os.listdir('../input/yikes-spiders-15-species/test/' + classes[i] + '/')
    for j in range(len(image_path)):
        image = Image.open('../input/yikes-spiders-15-species/test/' + classes[i] + '/' + image_path[j])
        input_image = preprocess(image).unsqueeze(0)
        test.append([input_image, i])
```

```python
alexnet = models.alexnet(pretrained=True)
alexnet.eval()
```

```python
# random.shuffle(train)
# random.shuffle(test)
```

```python
for i in range(len(train)):
    train[i][0] = alexnet(train[i][0]).detach().numpy()
    if(i % 100 == 0):
        print(i)
    
for i in range(len(test)):
    test[i][0] = alexnet(test[i][0]).detach().numpy()
    if(i % 100 == 0):
        print(i)
```

```python
x_train = []
y_train = []
x_test = []
y_test = []

for i in range(len(train)):
    x_train.append(train[i][0])
    y_train.append(train[i][1])

del train

for i in range(len(test)):
    x_test.append(test[i][0])
    y_test.append(test[i][1])

del test
```

```python
N, nx, ny = (np.array(x_train)).shape

x_train = np.array(x_train)

x_train = x_train.reshape((N, nx * ny))
```

```python
N, nx, ny = (np.array(x_test)).shape

x_test = np.array(x_test)

x_test = x_test.reshape((N, nx * ny))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

#Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression(max_iter=3000)
```

```python
model.fit(x_train, y_train)
yhat = model.predict(x_test)
```

```python
print(accuracy_score(yhat, y_test))
```

**SVM Classifier**

```python
from sklearn import svm
model = svm.SVC()
```

```python
model.fit(x_train, y_train)
yhat = model.predict(x_test)
```

```python
print(accuracy_score(yhat, y_test))
```

**GNB Classifier**

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train, y_train)
yhat = model.predict(x_test)
```

```python
print(accuracy_score(yhat, y_test))
```

**Bike vs Horse Classification**

```python
image_dataset = []   # 0 for bikes and 1 for horses.
categories = []
for image in os.listdir('../input/bikeshorses/Assignment2_BikeHorses/Bikes'):
    image_dataset.append(preprocess(Image.open('../input/bikeshorses/Assignment2_BikeHorses/Bikes/' + image)).unsqueeze(0))
    categories.append(0)
    
for image in os.listdir('../input/bikeshorses/Assignment2_BikeHorses/Horses'):
    image_dataset.append(preprocess(Image.open('../input/bikeshorses/Assignment2_BikeHorses/Horses/' + image)).unsqueeze(0))
    categories.append(1)
```

```python
alexnet = models.alexnet(pretrained=True)
alexnet.eval()
```

```python
for i in range(len(image_dataset)):
    image_dataset[i] = alexnet(image_dataset[i]).detach().numpy()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(image_dataset, categories, test_size = 0.25, random_state = 42)
```

```python
N, nx, ny = (np.array(x_train)).shape

x_train = np.array(x_train)

x_train = x_train.reshape((N, nx * ny))
```

```python
N, nx, ny = (np.array(x_test)).shape

x_test = np.array(x_test)

x_test = x_test.reshape((N, nx * ny))
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
yhat = model.predict(x_test)
```

```python
print(accuracy_score(yhat, y_test))
```

**SVM Classifier**

```python
from sklearn import svm
model = svm.SVC()
```

```python
model.fit(x_train, y_train)
yhat = model.predict(x_test)
```

```python
print(accuracy_score(yhat, y_test))
```

**GNB Classifier**

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train, y_train)
yhat = model.predict(x_test)
```

```python
print(accuracy_score(yhat, y_test))
```
